```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = pd.read_csv('spam_ham_dataset.csv')
data.head()
```

```python

df = data.drop(data.columns[3],axis=1)
df.head()
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.head()
main_data=df
main_data.head()
```

```python
main_data.shape
```

```python
main_data.loc[main_data['label'] == 'spam', 'lable',] = 0
main_data.loc[main_data['label'] == 'ham', 'lable',] = 1
main_data.head()
```

```python
main_data.shape
```

```python
X=main_data['text']
X
```

```python
Y=main_data['lable']
Y
```

model creation

```python

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=43)

# Create a TF-IDF vectorizer
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

# Transform the training and testing data
X_train_feature = feature_extraction.fit_transform(X_train)
X_test_feature = feature_extraction.transform(X_test)

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
#model creation
model_svm = SVC(kernel='linear', C=1.0)  # You can choose different kernels and hyperparameters
model_svm.fit(X_train_feature, Y_train)

prediction = model_svm.predict(X_test_feature)

# Evaluate the model's performance
accuracy = accuracy_score(Y_test,prediction)

report= classification_report(Y_test,prediction)
```

```python
print("accuracy:{2f}",accuracy)
```

```python
print(report)
```
